```python
import pandas as pd
import pyam
import numpy as np
```

```python
equity_data = pd.read_csv(
    '../data/figure3b.csv',
    index_col=[0,1]
)
```

First, reset index for equity data, group by model and scenario and sum up the following columns:
* Carbon Dioxide Removal|NovelCmltv
* Carbon Dioxide Removal|Novel|FairShare
* Carbon Dioxide Removal|Novel [p90]|Additional|FairShare

```python
cols_to_sum = [
    'Carbon Dioxide Removal|Novel|FairShare',
    'Carbon Dioxide Removal|Novel [p90]|Additional|FairShare'
]
```

```python
summary_equity = (
    equity_data
    .reset_index()
    .groupby(['Model', 'Scenario'])[cols_to_sum]
    .sum()
    .round(0)
)
```

```python
summary_equity
```

```python
global_data_small = global_data.loc[summary_equity.index, :].round(0)
```

```python
for ix in global_data_small.index:
    assert global_data_small.loc[ix,'p90'] == summary_equity.loc[ix,'Carbon Dioxide Removal|Novel [p90]|Additional|FairShare']
```

Ok, we don't have the novel CDR starting from 2020. Let us calculate this quickly.

```python
novel_cdr = pd.read_csv(
    '../data/100_novel_cdr.csv',
    index_col=[0,1]
)
```

```python
novel_cdr_select = pyam.IamDataFrame(novel_cdr.loc[summary_equity.index, :].reset_index())
```

```python
novel_cdr_select_cum = (
    novel_cdr_select
    .timeseries()
    .apply(
        lambda x: pyam.timeseries.cumulative(
            x,
            first_year=2020,
            last_year=2100
        ),
        axis=1
    )
)
```

```python
novel_cdr_select_cum /= 1e3
```

```python
novel_cdr_select_cum.round(0)
```
